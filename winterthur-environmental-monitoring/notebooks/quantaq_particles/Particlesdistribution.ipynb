```python
bin_cols = [f"bin{i}" for i in range(21)] 
```

```python
avg_1555 = df_1555_clean[bin_cols].mean()
avg_1557 = df_1557_clean[bin_cols].mean()
```

```python
norm_1555 = avg_1555 / avg_1555.sum()
norm_1557 = avg_1557 / avg_1557.sum()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(bin_cols, avg_1555, marker='o', label='1555 (China Hall)')
plt.plot(bin_cols, avg_1557, marker='s', label='1557 (Conservatory)')
plt.xlabel("Particle Size Bin")
plt.ylabel("Average Count")
plt.title("Average Particle Count per Bin (Raw Data, Sept 2025)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(bin_cols, norm_1555, marker='o', label='1555 (Normalized)')
plt.plot(bin_cols, norm_1557, marker='s', label='1557 (Normalized)')
plt.xlabel("Particle Size Bin")
plt.ylabel("Fraction of Total Counts")
plt.title("Normalized Particle Size Distribution (Sept 2025)")
plt.legend()
plt.grid(True)
plt.show()
```
